### COVID 19 data clean up and plotting

```python
# import Dependencies
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
# Get datafile
allState_path = "all-states-history.csv"
usCovid_path = "national-history.csv"
# Read the data and assign variable
allState_covid = pd.read_csv(allState_path)
usCovid_covid = pd.read_csv(usCovid_path)
```

```python
# Display to Analyze and clean up dataframe for the US
usCovid_covid
```

```python
# Make new data frame containing only two columns
usCovid_df = usCovid_covid.loc[:,['date', 'positive']]
usCovid_df
```

```python
# Sort data frame so that numbers of positive cases will be smalles on top
#cov = usCovid_df.sort_values(by=['positive'])
#cov.head(20)
```

```python
# Set x axis 
x_axis = np.arange(len(usCovid_df))
```

```python
# Create a list indicating where to write x labels and set figure size to adjust for space
f = plt.figure(figsize=(30,5))
ax = f.add_subplot(111)
# Combine line plot and bar plot to one plot
plt.plot(x_axis, usCovid_df['positive'])
plt.bar(x_axis, usCovid_df['positive'], color = 'r', alpha = 0.50,align = 'edge')
#plt.xticks(tick_locations)
ax.set_xticks([11,41,72,102,133,164,194,224,255,285,314,325])
ax.set_xticklabels(["Dec 20'","Nov 20'", "Oct 20'", "Sept 20'", "Aug 20'", "Jul 20'",
                   "Jun 20'", "May 20'", "Apr 20'","Mar 20'", " Feb 20'", "Jan 20'"])
plt.xlim(0,325)
ax.get_yaxis().set_major_formatter(matplotlib.ticker.FuncFormatter(lambda x, p: format(int(x), ',')))
# Save image in directory COVID_data
plt.savefig('covid_us_barplot.png')
plt.show()
```
